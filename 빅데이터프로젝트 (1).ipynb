```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np


plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
plt.style.use('ggplot')

# ----------------------------------------------------
# 1. 데이터 로드 및 전처리
# ----------------------------------------------------
print("--- 1. 데이터 로드 및 전처리 시작 ---")
file_name = "빅데이터이음.csv" # 파일 이름 변수화 (사용자 파일 이름)


try:
    df_data = pd.read_csv(file_name, encoding='utf-8')
except UnicodeDecodeError:
        exit()

# '일자' 컬럼을 datetime으로 변환하고 '월', '군구' 컬럼 생성
df_data['일자'] = pd.to_datetime(df_data['일자'])
df_data['월'] = df_data['일자'].dt.to_period('M').astype(str)
df_data['군구'] = df_data['지역'].apply(lambda x: x.split()[-1])

print("데이터 로드 및 전처리 완료.")

# ----------------------------------------------------
# 2. 분석 및 요약 데이터 생성
# ----------------------------------------------------
print("\n--- 2. 분석 및 요약 데이터 생성 시작 ---")
industry_cols = ['문화취미-영화관', '숙박', '약국', '연료', '유통업영리 - 편의점', '일반휴게', '학원', '기타']
group_cols = ['전체금액'] + industry_cols

# 2-1. 업종별 총 소비 금액 분석 (인천시 업종 별 소비 금액.csv)
df_industry_total = df_data[industry_cols].sum().reset_index()
df_industry_total.columns = ['업종', '총 소비 금액']
df_industry_total = df_industry_total.sort_values(by='총 소비 금액', ascending=False)
df_industry_total['총 소비 금액 (억 원)'] = (df_industry_total['총 소비 금액'] / 100000000).round(2)
df_industry_total.to_csv("인천시 업종 별 소비 금액.csv", index=False, encoding='utf-8-sig')

# 2-2. 군/구별 총 소비 금액 요약 ( 군구 별 소비 금액.csv)
df_district_summary = df_data.groupby('군구')[group_cols].sum().reset_index()
df_district_summary = df_district_summary.sort_values(by='전체금액', ascending=False)
df_district_summary['전체금액 (억 원)'] = (df_district_summary['전체금액'] / 100000000).round(2)
df_district_summary.to_csv(" 군구 별 소비 금액.csv", index=False, encoding='utf-8-sig')

# 2-3. 일자별 총 소비 금액 추이 데이터 (result3.png에 사용)
df_daily_total = df_data.groupby(df_data['일자'].dt.date)['전체금액'].sum().reset_index()
df_daily_total.columns = ['일자', '일별 총 소비 금액']
df_daily_total['일별 총 소비 금액 (억 원)'] = df_daily_total['일별 총 소비 금액'] / 100000000

print("분석 및 요약 데이터 생성 완료. (2개 CSV 파일 저장)")

# ----------------------------------------------------
# 3. 데이터 시각화
# ----------------------------------------------------

print("\n--- 3. 데이터 시각화 시작 (4개 PNG 파일 생성) ---")

# 3-1. 업종별 총 소비 금액 시각화 (막대 그래프) -> result1.png
plt.figure(figsize=(12, 6))
sns.barplot(
    x='업종',
    y='총 소비 금액 (억 원)',
    data=df_industry_total,
    palette='viridis'
)
# 제목 수정: 2025년 1월
plt.title('업종별 총 소비 금액 (2025년 1월)', fontsize=16)
plt.xlabel('업종', fontsize=12)
plt.ylabel('총 소비 금액 (억 원)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('result1.png')
plt.close()

# 3-2. 군/구별 총 소비 금액 시각화 (막대 그래프) -> result2.png
plt.figure(figsize=(12, 6))
sns.barplot(
    x='군구',
    y='전체금액 (억 원)',
    data=df_district_summary,
    palette='magma'
)
# 제목 수정: 2025년 1월
plt.title('군/구별 총 소비 금액 (2025년 1월)', fontsize=16)
plt.xlabel('군/구', fontsize=12)
plt.ylabel('총 소비 금액 (억 원)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('result2.png')
plt.close()

# 3-3. 25.1월 일자별 총 소비 금액 그래프 (선 그래프) -> result3.png
plt.figure(figsize=(14, 7))
sns.lineplot(
    x='일자',
    y='일별 총 소비 금액 (억 원)',
    data=df_daily_total,
    marker='o',
    color='darkorange',
    linewidth=2
)
# ⭐️ 제목 수정: 2025년 1월
plt.title('2025년 1월 일자별 총 소비 금액 추이', fontsize=16)
plt.xlabel('일자', fontsize=12)
plt.ylabel('일별 총 소비 금액 (억 원)', fontsize=12)
plt.xticks(df_daily_total['일자'], rotation=45, ha='right')
plt.grid(axis='both', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('result3.png')
plt.close()

# ⭐️ 3-4. 군/구별 업종별 소비 금액 비중 시각화  -> result4.png
print("\n--- 3-4. 군/구별 업종별 소비 비중 (Stacked Bar Plot) 생성 ---")


# '군구'를 인덱스로 설정하고 업종별 금액만 선택
df_plot = df_district_summary.set_index('군구')[industry_cols]

# 금액 단위를 '억 원'으로 변환하여 시각화 가독성 높이기
df_plot_billion = df_plot / 100000000

# Pandas의 plot 메서드를 사용하여 생성
plt.figure(figsize=(15, 8))
ax = df_plot_billion.plot(
    kind='bar',
    stacked=True,
    figsize=(15, 8),
    colormap='Spectral',
    edgecolor='black',
    ax=plt.gca()
)

# : 2025년 1월
plt.title('군/구별 업종별 소비 금액 비중 (2025년 1월)', fontsize=16)
plt.xlabel('군/구', fontsize=12)
plt.ylabel('총 소비 금액 (억 원)', fontsize=12)
plt.xticks(rotation=45, ha='right')

# 범례 위치를 그래프 밖으로 이동하여 겹치지 않게 설정
plt.legend(title='업종', bbox_to_anchor=(1.05, 1), loc='upper left')

# 그래프 저장
plt.tight_layout(rect=[0, 0, 0.85, 1]) # 범례를 위한 공간 확보
plt.savefig('result4.png')
plt.close()

print("모든 시각화 파일 생성이 완료되었습니다: result1.png, result2.png, result3.png, result4.png")

print("\n--- 전체 코드 실행 완료 ---")
```
